```python
%matplotlib inline
from reconstruction.data import CV_Wigner
import matplotlib.pyplot as plt
import qutip as qp
import numpy as np

state = (qp.fock(5, 1) - 1j*qp.fock(5, 4)).unit()

exp = CV_Wigner()
exp.simulateData(state = state,noise_amp = 5e-3)
exp.plotData()
exp.calcDesignMatrix(5)
```

```python
rho = np.empty( (5, 5) )

for idx in np.arange(5):
    for idy in np.arange(5):
        c_op = 0.5*(qp.fock(5, idx) * (qp.fock(5, idy)).dag() 
                        + qp.fock(5, idy) * (qp.fock(5, idx)).dag() ) + 0.5j * (
                    qp.fock(5, idx) * (qp.fock(5, idy)).dag()
                        - qp.fock(5, idy) * (qp.fock(5, idx)).dag() )
        rho[idx, idy] = exp.getObs(c_op)
print rho
qp.hinton(rho)
```
