```python
#Create ideal bell-cat and bell states
N = 15
disp = np.sqrt(3.0)
exp.dispAmp = disp
state1 = qp.tensor( qp.fock(2,0), qp.coherent(N, disp)  )
state2 = qp.tensor( qp.fock(2,1), qp.coherent(N, -disp) )
state = (state1 + state2).unit()

state_dm = state*state.dag()

stateA = qp.tensor( qp.fock(2,0), qp.fock(2, 0)  )
stateB = qp.tensor( qp.fock(2,1), qp.fock(2, 1)  )
stateL = (stateA + stateB).unit()

state_dm = state*state.dag()
```

```python
q_ops, c_ops = exp.makeOps()
exp.reduceQubit()

norm = exp.rho_reduced.tr()

p_ideal = vis.paulibar_2q(stateL*stateL.dag())
p_norm = vis.paulibar_2q(exp.rho_reduced/norm)
p_totn = vis.paulibar_2q(exp.rho_reduced/np.trace(exp.density_matrix_inv))
p_raw = vis.paulibar_2q(exp.rho_reduced)
```
