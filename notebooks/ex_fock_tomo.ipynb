```python
probs = np.real( blah.density_matrix_min.diagonal() )
n = range(len(probs))

fig = plt.figure()
ax = fig.add_subplot(111)
ax.bar(n, probs[n])
ax.set_title('Reconstructed photon number probability')
ax.set_xlabel('Photon number')
ax.set_ylabel('Probability')

qp.hinton(blah.density_matrix_min)

no_fock_probs = probs
```

```python
fock_counts = 6.3e5
no_fock_counts = 2.4e6
combined = (fock_counts * fock_probs + no_fock_counts * no_fock_probs )  / (fock_counts + no_fock_counts) 

fig = plt.figure(figsize = (10,4))
ax = fig.add_subplot(121)
ax.bar(n, combined[n])
ax.set_title('Reconstructed')
ax.set_xlabel('Photon number')
ax.set_ylabel('Probability')

N = 10
ideal = qp.coherent_dm(N, np.sqrt(3)).full().diagonal()
ax = fig.add_subplot(122)
ax.bar(n, ideal[n])
ax.set_title('Ideal')
ax.set_xlabel('Photon number')
ax.set_ylabel('Probability')
```
