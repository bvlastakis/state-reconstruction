```python
state = qp.fock(6,0) #+ qp.fock(6,3)
#state = qp.fock(6,0)
#state = qp.fock_dm(6,0) + qp.fock_dm(6,3) + 0*(qp.fock(6,0)*qp.fock(6,3).dag() + qp.fock(6,3)*qp.fock(6,0).dag())
#state = qp.coherent(25,2)
state = state.unit()

dispVec = np.linspace(-3,3,21)
mesh_Real, mesh_Imag = np.meshgrid(dispVec, dispVec)
dispGrid = mesh_Real + 1j * mesh_Imag

noise_amp = 0.05
path = "sample_data/XE.txt"

blah = dt.CV_Wigner()
#blah.importData(path)
blah.simulateData(state.unit(), noise_amp, dispGrid)
blah.plotData()
```

```python
basis = 7

blah.calcDesignMatrix(basis)
blah.regression(method = 'inversion')
blah.plotWigner(blah.density_matrix_inv, factor = 10, 
                title = 'Reconstructed Wigner function')
#blah.plotDiagonal(blah.density_matrix_min)
```

```python
N_regress = 20
A = pd.DataFrame([])

for i in arange(N_regress):
    blah.simulateData(state.unit(), noise_amp, dispGrid)
    dM = blah.regression(method = 'inversion')
    D = pd.DataFrame(np.real(dM.reshape( (1,basis**2) )) )
    A = pd.concat( (A, D) , ignore_index = True )
```

```python
fig = plt.figure(figsize(16,8))
bp = A.boxplot()

dd = np.arange(basis**2).reshape((basis,basis))
diag = dd.diagonal()

photon_prob = A.ix[:,diag]

fig = plt.figure(figsize(16,8))
bp = photon_prob.boxplot()
```
