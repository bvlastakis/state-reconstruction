Here we will make a simulated data set use a multiple nonlinear regression to perform a density matrix reconstruction.

```python
%pylab inline

import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


import qutip as qp
import reconstruction.data as dt

reload(dt)
```

```python
#path = "sample_data/XE.txt"

#state = qp.coherent(20,2) - qp.coherent(20,-2) + qp.coherent(20,2j) - qp.coherent(20,-2j)
state = qp.fock(6,0) + qp.fock(6,3)
#state = qp.fock_dm(6,0) + qp.fock_dm(6,3) + 0.5*(qp.fock(6,0)*qp.fock(6,3).dag() + qp.fock(6,3)*qp.fock(6,0).dag())

dispVec = np.linspace(-3,3,41)
mesh_Real, mesh_Imag = np.meshgrid(dispVec, dispVec)
dispGrid = mesh_Real + 1j * mesh_Imag

basis = 10

noise_amp = 0.05

blah = dt.CV_Wigner()
blah.simulateData(state.unit() ,noise_amp, dispGrid)
blah.plotData()
blah.calcDesignMatrix(basis)
blah.regression(method = 'inversion')
blah.plotWigner(blah.density_matrix_inv, factor = 10, title = 'Reconstructed Wigner function')
```

```python
probs = np.real( blah.density_matrix_inv.diagonal() )
n = range(len(probs))

fig = plt.figure()
ax = fig.add_subplot(111)
ax.bar(n, probs[n])
ax.set_title('Reconstructed photon number probability')
ax.set_xlabel('Photon number')
ax.set_ylabel('Probability')
```

```python
import cvxpy as cp
import cvxopt
import numpy as np


n = basis #number of elements in density matrix

W1 = blah.param_vec
M1 = blah.dM_complex

#W2 = np.ones(W1.shape)
#M2 = np.ones(M1.shape)

#W = cvxopt.matrix(W1)
#M = cvxopt.matrix(M1)
x = cp.SDPVar(2*n,2*n)

y = cp.Variable(0)
for idx in np.arange(x.size[1]/2):
    y = cp.vstack(y, x[:x.size[1]/2, idx])
for idx in (x.size[1]/2 + np.arange(x.size[1]/2)):
    y = cp.vstack(y, x[:x.size[1]/2, idx])
#y = cp.Variable(2*n**2)

objective = cp.Minimize( cp.sum( cp.square(M1*y - W1) ) )
constraints = [cp.lambda_min(x) >= 0]

p = cp.Problem(objective, constraints)

result = p.solve()
# The optimal value for x is stored in x.value.
```

```python
tolerance = 1e-6
A = np.array(x.value)
A[np.abs(A) < tolerance] = 0

B = A[:A.shape[1]/2, :A.shape[1]/2]
C = A[A.shape[1]/2:, :A.shape[1]/2]

density = B + 1j*C

blah.plotWigner(density, title = 'Reconstructed Wigner function')

probs = np.real( density.diagonal() )
n = range(len(probs))

fig = plt.figure()
ax = fig.add_subplot(111)
ax.bar(n, probs[n])
ax.set_title('Reconstructed photon number probability')
ax.set_xlabel('Photon number')
ax.set_ylabel('Probability')
```
